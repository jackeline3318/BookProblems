```python
import pandas as pd
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
import numpy as np

A = pd.read_csv(r"C:\Users\gisse\Downloads\BlueOrRed.csv")

X = A[['Age', 'HouseholdSize', 'Income', 'Education']]
y = A['Undecided']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

def find_optimal_k(X_train, y_train, max_k=20):
    error_rates = []
    k_values = list(range(1, max_k + 1))

    for k in k_values:
        knn = KNeighborsClassifier(n_neighbors=k)
        cv_scores = cross_val_score(knn, X_train, y_train, cv=10, scoring='accuracy')
        error_rate = 1 - np.mean(cv_scores)
        error_rates.append(error_rate)

    optimal_k = k_values[np.argmin(error_rates)]
    return optimal_k

optimal_k = find_optimal_k(X_train, y_train)

final_knn_model = KNeighborsClassifier(n_neighbors=optimal_k)
final_knn_model.fit(X_train, y_train)

accuracy = final_knn_model.score(X_val, y_val)
error_rate = 1 - accuracy

print(f"Optimal k: {optimal_k}")
print(f"Validation Set Accuracy: {accuracy:.4f}")
print(f"Validation Set Error Rate: {error_rate:.4f}")
```

```python
import pandas as pd
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
import numpy as np

A = pd.read_csv(r"C:\Users\gisse\Downloads\BlueOrRed.csv")

X = A.drop('Undecided', axis=1)
y = A['Undecided']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

def find_optimal_k(X_train, y_train, max_k=20):
    error_rates = []
    k_values = list(range(1, max_k + 1))

    for k in k_values:
        knn = KNeighborsClassifier(n_neighbors=k)
        cv_scores = cross_val_score(knn, X_train, y_train, cv=10, scoring='accuracy')
        error_rate = 1 - np.mean(cv_scores)
        error_rates.append(error_rate)

    optimal_k = k_values[np.argmin(error_rates)]
    return optimal_k

optimal_k = find_optimal_k(X_train, y_train)

final_knn_model = KNeighborsClassifier(n_neighbors=optimal_k)
final_knn_model.fit(X_train, y_train)

accuracy = final_knn_model.score(X_val, y_val)
error_rate = 1 - accuracy

print(f"Optimal k: {optimal_k}")
print(f"Validation Set Accuracy: {accuracy:.4f}")
print(f"Validation Set Error Rate: {error_rate:.4f}")
```
